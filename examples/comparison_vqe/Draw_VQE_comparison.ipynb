```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
qiskit = np.load("vqe_qiskit.npz")
pennylane = np.load("vqe_pennylane.npz")
tedq = np.load("vqe_tedq.npz")
```

```python
font = {'family' : 'normal',
        'size'   : 12}

matplotlib.rc('font', **font)
plt.plot(tedq["t"], tedq["e"], 'r-', label="TeD-Q")
plt.plot(pennylane["t"], pennylane["e"], '-', color="orange", label="Pennylane")
plt.plot(qiskit["t"], qiskit["e"], 'y-', label="Qiskit")

plt.xlabel('Time(s)')
plt.grid()
plt.ylabel('Error')
plt.gca().set_yscale('log')
plt.legend()

plt.savefig("vqe_error_time_comparison.jpg")
```

```python
matplotlib.rc('font', **font)

plt.plot(tedq["d"], tedq["el"], 'r-', label="TeD-Q", linewidth=4)
plt.plot(pennylane["d"], pennylane["el"], '-', color="orange", label="Pennylane")
plt.plot(qiskit["d"], qiskit["el"], 'y-', label="Qiskit")
plt.plot(tedq["d"], tedq["rel"] , '--', color="black", label="FCI reference energy")
plt.xlabel('Distance(a0)')
plt.grid()
plt.ylabel('Energy(Hartree)')

plt.legend()

plt.savefig("vqe_error_time_comparison.jpg")
```

```python
matplotlib.rc('font', **font)

plt.plot(tedq["d"], tedq["el1"], 'r-', label="TeD-Q", linewidth=4)
plt.plot(pennylane["d"], pennylane["el1"], '-', color="orange", label="Pennylane")
plt.plot(qiskit["d"], qiskit["el1"], 'y-', label="Qiskit")
plt.plot(tedq["d"], tedq["rel"] , '--', color="black", label="FCI reference energy")
plt.xlabel('Distance(a0)')
plt.grid()
plt.ylabel('Energy(Hartree)')

plt.legend()

plt.savefig("vqe_error_time_comparison.jpg")
```
