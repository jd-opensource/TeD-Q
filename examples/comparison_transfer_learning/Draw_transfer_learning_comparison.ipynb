```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
qiskit = np.load("transfer_learning_qiskit.npz")
pennylane = np.load("transfer_learning_pennylane.npz")
tedq = np.load("transfer_learning_tedq.npz")
```

```python
font = {'family' : 'normal',
        'size'   : 12}

matplotlib.rc('font', **font)
plt.plot(tedq["t"], tedq["a"], 'r-', label="TeD-Q")
plt.plot(pennylane["t"], pennylane["a"], '-', color="orange", label="Pennylane")
plt.plot(qiskit["t"], qiskit["a"], 'y-', label="Qiskit")
# plt.gca().set_xscale('log')

plt.xlabel('Time(s)')
plt.grid()
plt.ylabel('Accuracy(%)')

plt.legend()

plt.savefig("transfer_learning_comparison.jpg")
```
