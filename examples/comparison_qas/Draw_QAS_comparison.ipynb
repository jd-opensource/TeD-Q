```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
pennylane = np.load("./qas_pennylane.npz")
tedq = np.load("./qas_tedq.npz")
```

```python
font = {'family' : 'normal',
        'size'   : 12}

matplotlib.rc('font', **font)
plt.plot(tedq["t"] , 'r-', label="TeD-Q")
plt.plot(pennylane["t"], '-', color="orange", label="Pennylane")

plt.ylabel('Time (s)')
plt.grid()
plt.xlabel('Iteration (#)')
plt.legend()

plt.savefig("vqe_error_time_comparison.jpg")
```

```python
font = {'family' : 'normal',
        'size'   : 12}

matplotlib.rc('font', **font)
plt.plot(tedq["r"] , 'r-', label="TeD-Q")

plt.ylabel('Energy (Ha)')
plt.grid()
plt.xlabel('Iteration (#)')
plt.legend()

plt.savefig("vqe_error_time_comparison.jpg")
```
